```python
dogimg = cv2.imread('doggo.jpg')

font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(dogimg, text='I', org=(30, 40), fontFace=font, fontScale=0.7, color=(255, 255, 255), thickness=1, lineType=cv2.LINE_AA)
cv2.putText(dogimg, text='Dog', org=(10, 100), fontFace=font, fontScale=0.7, color=(255, 255, 255), thickness=1, lineType=cv2.LINE_AA)

pts = np.array([[18,64], [36,83], [53,64]])
print(pts.shape)
pts = pts.reshape((-1, 1, 2))
print(pts.shape)
cv2.fillPoly(dogimg, [pts], color=(255, 20, 147))

cv2.circle(dogimg, center=(27, 60), radius=10, color=(255, 20, 147), thickness=-1)
cv2.circle(dogimg, center=(45, 60 ), radius=10, color=(255, 20, 147), thickness=-1)
plt.imshow(dogimg)
```
